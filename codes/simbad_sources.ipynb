### Mass Calculation

```python
c = 3.0E+8      # speed of light in vacuum
h = 6.62E-34    # Planck's constant
kB = 1.38E-23   # Boltzmann constant
lam_iras = 10.0E-3    # Long-Wavelength IRAS
lam_akari = 1.4E-4    # Long-Wavelength Akari
lam_wise = 2.2E-5
```

```python
def Mass(lam, Temp, flux, Dis=979.4899287650741):
    def planck(lam):
        a = np.exp(h*c/(kB*lam*Temp))
        b = 1/(a-1)
        Bv = h*c*b/lam**3     # Planck distribution
        return Bv    
    Su = flux*5.288*10**-29   # Absolute Flux 
    D = Dis*3.0857*10**16     # D in Km, Dis in pc
    M = 0.4*Su*D**2/planck(lam)
    return M
```

```python
M_iras = Mass(lam_iras, iras.Temp, iras.F100)
iras['Mass'] = M_iras
np.mean(M_iras/(2E+30))   #in solar mass
```

```python
M_akari = Mass(lam_akari, akari.Temp, akari.F140)
akari['Mass'] = M_akari
np.mean(M_akari/(2E+30))   #in solar mass
```

```python
M_wise = Mass(lam_wise, wise.Temp, wise.F22)
wise['Mass'] = M_wise
np.mean(M_wise/(2E+30))
```

```python
def label_and_title(title='IRIS', xlim=[3.55, 4.05], ylim=[-39.3, -39.1]):
    fontsize=20
    plt.title(title, weight='bold', fontsize=fontsize)
    plt.xlabel('RA (deg.)', fontsize = fontsize, weight='bold')
    plt.ylabel('DEC (deg.)', fontsize = fontsize, weight='bold')
#     plt.xticks(fontsize=fontsize, weight='bold')
#     plt.yticks(fontsize=fontsize, weight='bold')
    plt.xlim(xlim)
    plt.ylim(ylim)
```
