```python
def scatter(data1, data2):
    coeff = np.polyfit(data1, data2, deg=1)
    p = np.poly1d(coeff)
    y = p(data1)
    plt.plot(data1, y, label='Straight Line Fit', lw=3, color='black')
    plt.scatter(data1, data2, color='m', label='data')
    plt.legend()
```

## GAUSSIAN FIT

```python
def gaussian(x, nbins, xlabel= 'Temp (K)', fontsize=30):
    mu = np.mean(x)
    sigma = np.std(x)
    n, bins, patches = plt.hist(x, nbins, density=True, color="white", linewidth=3, edgecolor='k')
    y = ((1/(np.sqrt(2 * np.pi) * sigma))*np.exp(-0.5 * (1/sigma*(bins - mu))**2))
    plt.plot(bins, y, lw=5)
    plt.xlabel(xlabel, fontsize=fontsize, weight='bold')
    plt.ylabel('Probability Density', fontsize=fontsize, weight='bold')
    plt.xticks(fontsize=fontsize, color = 'k')
    plt.yticks(fontsize=fontsize,color='k')
    plt.tight_layout()
```

```python
num_bins = 40
fontsize=30
plt.figure(figsize=(45,12))
plt.subplot(131)
plt.title('IRIS', fontweight = 'bold', fontsize=fontsize)
# plt.xlim([23,34])
gaussian(data_iras.Temp, num_bins)

plt.subplot(132)
plt.title('AKARI', fontweight = 'bold', fontsize=fontsize)
# plt.xlim([13,29])
gaussian(data_akari.Temp, num_bins)

plt.subplot(133)
plt.title('WISE', fontweight = 'bold', fontsize=fontsize)
# plt.xlim([380,460])
gaussian(data_wise.Temp, num_bins)

# plt.savefig(FIG_DIR+"tempGaussian.png")
```

## MASS CALCULATION

```python
c = 3.0E+8      # speed of light in vacuum
h = 6.62E-34    # Planck's constant
kB = 1.38E-23   # Boltzmann constant
lam_iras = 10.0E-3    # Long-Wavelength IRAS
lam_akari = 1.4E-4    # Long-Wavelength Akari
lam_wise = 2.2E-5
```

```python
def Mass(lam, Temp, Dis, flux):
    def planck(lam):
        a = np.exp(h*c/(kB*lam*Temp))
        b = 1/(a-1)
        Bv = h*c*b/lam**3     # Planck distribution
        return Bv    
    Su = flux*5.288*10**-29   # Absolute Flux 
    D = Dis*3.0857*10**16     # D in Km, Dis in pc
    M = 0.4*Su*D**2/planck(lam)
    return M
```

```python
M_iras = Mass(lam_iras, data_iras.Temp, Dis, data_iras.F100)
data_iras['Mass'] = M_iras
np.mean(M_iras/(2E+30))   #in solar mass
```

```python
M_akari = Mass(lam_akari, data_akari.Temp, Dis, data_akari.F140)
data_akari['Mass'] = M_akari
np.mean(M_akari/(2E+30))   #in solar mass
```

```python
M_wise = Mass(lam_wise, data_wise.Temp, Dis, data_wise.F22)
data_wise['Mass'] = M_wise
np.mean(M_wise/(2E+30))
```

## Visual Extinction

```python
scale = 5.288e-29
def extinction(lam, Temp, flux):
    a = np.exp(h*c/(kB*lam*Temp))
    b = 1/(a-1)
    planck = h*c*b/lam**3
    f = flux*scale
    opt_depth = (f/planck)
    ext = 15.078*(1 - np.exp(-opt_depth/641.3))
    return ext
```

```python
# size of cavity = 23.184pc x 11.97pc
def inclination(a, b):
    axial_ratio = b/a  # b/a
    int_flat = 0.23 # (for now) 
    ratio = (axial_ratio**2-int_flat**2)/(1-int_flat**2)
    inc_angle = np.arccos(np.sqrt(ratio))*180/np.pi   # in degrees
    return inc_angle
```

```python
inc_iras = inclination(a_iras, b_iras)
inc_akari = inclination(a_akari, b_akari)
inc_wise = inclination(a_wise, b_wise)
inc_iras, inc_akari, inc_wise
```

## JEANS MASS

### $\rho = (\frac{3}{4\pi})^{2/3} \frac{k_b T}{m_HGR^2}$ 
### $M_J =  (\frac{k_b T}{m_HG})^{3/2}\frac{1}{\rho^{1/2}} $ 
#### $m_H$ = mass of Hydrogen

```python
def Jeans_mass(T, a = a, b = b, M_H = 1.67e-27, G = 6.67e-11):
    R = (a+b)/4 *3.0857e16    #mean diameter = (a+b)/2 so mean Radius = (a+b)/4
    density = (3/(4*np.pi))**(2/3)*kB*T/(M_H*G*R**2)
    M_Jeans = (kB*T/(M_H*G))**(3/2)*1/np.sqrt(density)
    return M_Jeans
```

```python
M_Jeans_iras = Jeans_mass(data_iras.Temp)
M_Jeans_akari = Jeans_mass(data_akari.Temp)
M_Jeans_wise = Jeans_mass(data_wise.Temp)
np.mean(M_Jeans_iras), np.mean(M_Jeans_akari), np.mean(M_Jeans_wise)
```

## PolyFit

```python
from sklearn.metrics import r2_score
from sklearn import linear_model
```

```python
def polyfit(x, y, xLabel, yLabel, title="IRIS", degree=1, fontsize = 14):
    plt.figure(figsize = [7,5])
    mymodel = np.poly1d(np.polyfit(x, y,degree))
    myline = np.linspace(min(x), max(x), len(y))
    r2 = r2_score(y, mymodel(x))
    plt.plot(x, y,'.b', label = 'Data')
    plt.plot(myline, mymodel(myline),'-',color = 'r',lw =3, label = 'Best Fit Polynomial, r$^2$='+str(np.round(r2, 2)))
    plt.title(title, weight="bold")
    plt.xlabel(xLabel, weight = 'bold',fontsize=fontsize)
    plt.ylabel(yLabel, weight = 'bold',fontsize=fontsize)
    plt.xticks(fontweight = None,fontsize=fontsize, color = 'k')
    plt.yticks(fontweight = None,fontsize=fontsize,color='k')
    plt.legend(fontsize = 10, loc = 'upper left')
#     print(r2_score(y, mymodel(x)))
```
